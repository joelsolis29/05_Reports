```python
import numpy as np
import matplotlib.pyplot as plt

# Categorias por calidad de roca
categories = ['I', 'II', 'III', 'IV', 'V']

# Valores por tipo de investigacin
estaciones_geomecanicas = [0, 40, 30, 30, 0]
perforaciones_diamantinas = [0, 60, 30, 10, 0]

# Posiciones de las barras
x = np.arange(len(categories))

# Ancho de barra
width = 0.4

# Crear grafico de barras
fig, ax = plt.subplots(figsize=(10, 4), dpi=480)

ax.bar(x - width/2, estaciones_geomecanicas, width, label='Estaciones geomecnicas', color='#778899')
ax.bar(x + width/2, perforaciones_diamantinas, width, label='Perforaciones diamantinas', color='#F5F5DC')

# Etiquetas y titulo
ax.set_xlabel('Calidad RMR')
ax.set_ylabel('Porcentaje (%)')
ax.set_title('RMR')
ax.set_axisbelow(True)
ax.grid(axis='y', color='k', linestyle='--', alpha=0.2)
ax.set_xticks(x)
ax.set_yticks(range(0,100,10))
ax.set_xticklabels(categories)

# Legenda
ax.legend()

# Mostrar grfico
plt.show()
```
