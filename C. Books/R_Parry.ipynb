## 1.1 Simple axial stress

Tengo un cilindro de radio y altura definidas, al cual lo someto a una fuerza de compresión $F$, este genera esfuerzos en los planos interiores del cilindro, los cuales van a variar según la inclinación $\theta$ del plano de falla. En base a eso determinaremos un fuerza normal ($N$) y de corte ($T$) en el plano mediante las siguientes ecuaciones:

$$N = F \cos \theta, T = F \sin \theta$$

Para convertir esto en esfuerzos, se divide entre el area elipsoidal generada por el plano, la cual es equivalente a $\frac{A}{\cos \theta}$

$$\sigma_{n \theta} = \frac{N \cos \theta}{A},\tau_{\theta} = \frac{T \cos \theta}{A}$$
$$\sigma_{n \theta} = \frac{(F \cos \theta) \cos \theta}{A},\tau_{\theta} = \frac{(F \sin \theta) \cos \theta}{A}$$
$$\sigma_{n \theta} = \frac{F \cos^2 \theta}{A},\tau_{\theta} = \frac{F \sin 2 \theta}{2A}$$

```python
import numpy as np
import math
class Cylinder():
    def __init__(self,diameter,length):
        self.d=diameter
        self.l=length
        self.area = 0.25*np.pi*(self.d**2)
    
    def stress_n(self, force, theta=0):
        sigma_n = force * (math.cos(theta))**2 /self.area
        tau_n = force * (math.sin(2*theta))/(2*self.area)
        return [sigma_n,tau_n]
```

```python
import matplotlib.pyplot as plt

# Creación del cilindro
cyl1 = Cylinder(0.05,50)
F = 150

# Rango de angulos
theta = np.linspace(0,np.pi,180)

# Calculo de esfuerzos
sigma_n = [cyl1.stress_n(F, t)[0] for t in theta]
tau_n = [cyl1.stress_n(F, t)[1] for t in theta]

# Grafico
plt.figure(figsize=(12,5), layout="tight")
plt.plot(theta, sigma_n)
plt.plot(theta, tau_n)
plt.xlabel(r'$\theta$ [rad]')
plt.ylabel(r'$\sigma_{n\theta}$ [Pa]')
plt.title('Esfuerzos vs. Ángulos de inclinación')
plt.grid(True)
plt.show()
```

## 1.2 Biaxial stress

Se extrae un cubo de lada $l$ un solido más grande, el cual se encuentra sometido a esfuerzos verticales y horizontales, por lo tanto, este cubo para un primer análisis se encuentra sometido unicamente a esfuerzos normales.

$$F_{1}=\sigma_{1} l, F_{2}=\sigma_{2} l \tan \theta$$

Los cuales se descomponen en fuerzas normales en el sentido del plano de inclinación $\theta$.

$$N_{1}+N_{2} = l \sigma_{n \theta} \sec \theta $$

$$F_{1} \cos \theta    + F_{2} \sin \theta = l \sigma_{n \theta} \sec \theta$$

$$\sigma_{1} l \cos \theta + \sigma_{2} l \tan \theta \sin \theta = l \sigma_{n \theta} \sec \theta$$

$$\sigma_{1} \cos^2 \theta + \sigma_{2} \sin^2 \theta = \sigma_{n \theta}$$

Y tambien se descomponenen en fuerzas cortantes:

$$T_{1}-T_{2} = l \tau_{n \theta} \sec \theta $$

$$F_{1} \sin \theta    - F_{2} \cos \theta = l \tau_{n \theta} \sec \theta$$

$$\sigma_{1} l \sin \theta - \sigma_{2} l \tan \theta \cos \theta = l \tau_{n \theta} \sec \theta$$

$$\sigma_{1} \sin 2 \theta - \sigma_{2} \sin 2 \theta = 2 \tau_{n \theta}$$

$$\frac{1}{2} (\sigma_{1}-\sigma_{2}) \sin 2 \theta= \tau_{n \theta}$$
